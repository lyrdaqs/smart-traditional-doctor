```python
import fasttext.util
model_skipgram = fasttext.load_model('drive/MyDrive/farsi-dedup-cbow.bin')
```

```python
model_skipgram.get_nearest_neighbors('آویشن', 40)
```

```python
def text_file_to_list(file_path):
    lines_list = []

    with open(file_path, 'r') as file:
        lines_list = [line.strip() for line in file]

    return lines_list

file_path1 = 'illness.txt'
illness = text_file_to_list(file_path1)

file_path2 = 'medicines.txt'
medicines = text_file_to_list(file_path2)
medicines
```

```python

with open("MedToSigns2.txt", "a", encoding="utf-8") as file:
     for medicine in medicines:
         similarities = []
         for sign in illness:
              distances = distance.cdist([model_skipgram.get_word_vector(medicine) - model_skipgram.get_word_vector("زعفران") + model_skipgram.get_word_vector("افسردگی")],
                                        [model_skipgram.get_word_vector(sign)], metric='cosine')[0]
              min_index = np.argmin(distances)
              min_distance = distances[min_index]
              max_similarity = 1 - min_distance
              similarities.append({"medicine": medicine, "sign": sign, "sim": max_similarity})

         top_similarities = sorted(similarities, key=lambda x: x['sim'], reverse=True)[:100]
         for similarity in top_similarities:
             output_line = f"Word1:{medicine},\tWord2:{similarity['sign']},\t\tSimularity:{similarity['sim']:.4f}\n"
             file.write(output_line)
             print(output_line, end="")
```

```python
with open("MedToIllness.txt", "a", encoding="utf-8") as file:
     for medicine in medicines:
         similarities = []
         for illnes in illness:
              distances = distance.cdist([model_skipgram.get_word_vector(medicine) - model_skipgram.get_word_vector("ریحان") + model_skipgram.get_word_vector("دیابت")],
                                        [model_skipgram.get_word_vector(illnes)], metric='cosine')[0]
              min_index = np.argmin(distances)
              min_distance = distances[min_index]
              max_similarity = 1 - min_distance
              similarities.append({"medicine": medicine, "illnes": illnes, "sim": max_similarity})

         top_similarities = sorted(similarities, key=lambda x: x['sim'], reverse=True)[:100]
         for similarity in top_similarities:
             output_line = f"{medicine},{similarity['illnes']},{similarity['sim']:.4f},ریحان دیابت\n"
             file.write(output_line)
             print(output_line, end="")
```
